```python
f = lambda t : exp(1j*t) + .5*exp(4.*1j*t)
```

```python
ts = linspace(0,2*pi,200)
u = real(f(ts))
v = imag(f(ts))
plot(u,v)
axis('equal');
```

```python
mask = u**2 > v**2

xres = 0*u;
resa = sign(u)*1.0*sqrt(u**2+v**2)
resb = sign(v)*u/v*sqrt(u**2+v**2)

xres[mask] = resa[mask]
xres[~mask] = resb[~mask]


yres = 0*u;
resa = sign(u)*v/u*sqrt(u**2+v**2)
resb = sign(v)*sqrt(u**2+v**2)

yres[mask] = resa[mask]
yres[~mask] = resb[~mask]
```

```python
plot(xres,yres)
```

```python
def discToSquare(u,v):
    mask = u**2 > v**2

    xres = 0*u;
    
    # This evaluates the different cases
    resa = sign(u)*1.0*sqrt(u**2+v**2)
    resb = sign(v)*u/v*sqrt(u**2+v**2)
    
    # This keeps only the relevant cases
    xres[mask] = resa[mask]
    xres[~mask] = resb[~mask]


    yres = 0*u;
    resa = sign(u)*v/u*sqrt(u**2+v**2)
    resb = sign(v)*sqrt(u**2+v**2)

    yres[mask] = resa[mask]
    yres[~mask] = resb[~mask]
    return (xres,yres)
```

```python
x,y = discToSquare(u,v)
plot(x,y)
axis('equal')
```

```python
f = lambda t : exp(4.*1j*t) + 1./2.*exp(7.*1j*t) + 1./4.*exp(10.*1j*t)
ts = linspace(0,2*pi,1000)
u = real(f(ts))
v = imag(f(ts))
subplot(1,2,1)
plot(u,v)
axis('equal')

subplot(1,2,2)
x,y = discToSquare(u,v);
plot(x,y)
axis('equal')
```
