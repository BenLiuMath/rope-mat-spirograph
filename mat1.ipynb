```python
ths = linspace(0,2*pi,1000);
```

```python
x = lambda t : sin(t) + .5*sin(4.*t)
y = lambda t : cos(t) + .5*cos(4.*t)

plot(x(ths),y(ths))
axis('equal')
axis(2.5*array([-1.,1.,-1.,1.]))
```

```python
def xfour(t) :
    return (72*sin(2*t - 1.35) +
        36*sin(4*t) +
        61*sin(5*t + 2.7) +
        20*sin(7*t + 1) +
        141*sin(8*t - 2.7) +
        19*sin(11*t + 1.35) +
        13*sin(14*t - 1) +
        15*sin(20*t - 2.7) +
        6*sin(26*t + 2.7))

def yfour(t) :
    return (72*cos(2*t - 1.35) +
        -36*cos(4*t) +
        61*cos(5*t + 2.7) +
        -20*cos(7*t + 1) +
        141*cos(8*t - 2.7) +
        19*cos(11*t + 1.35) +
        13*cos(14*t - 1) +
        15*cos(20*t - 2.7) +
        6*cos(26*t + 2.7))

plot(xfour(ths),yfour(ths))
axis('equal')
axis(300*array([-1.,1.,-1.,1.]))
```
